# 1. x + y

```python
def my_reward(x, y, theta, a):
    return abs(x+y)

plot_reward(my_reward)
```

```python
env = DrivingEnv(reward=my_reward)
model = StudentQModel(env)
model.train(10000, model_name="abs_xy");
```

# 2. Step

```python
def my_reward(x, y, theta, a):
    if x > 1.7:
        return 1.0
    else:
        return 0.0

plot_reward(my_reward)
```

```python
env.reward = my_reward

model = StudentQModel(env)
model.train(10000, model_name="step")
```

# 3. Negative Distance

```python
env = DrivingEnv()

def my_reward(x, y, theta, a):
    d, angle = env.distance_angle(x, y, theta)
    return -d

plot_reward(my_reward)
```

```python
env.reward = my_reward

model = StudentQModel(env)
model.train(50000, model_name="distance")
```
